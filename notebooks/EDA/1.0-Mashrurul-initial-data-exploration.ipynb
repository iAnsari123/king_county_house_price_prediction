### Data Exploration

**Column Names and descriptions for Kings County Data Set**
* **id** - unique identified for a house
* **date** - house was sold
* **price** -  Price of the house
* **bedroomsNumber** -  of Bedrooms/House
* **bathroomsNumber** -  of bathrooms/bedrooms
* **sqft_livingsquare** -  footage of the home
* **sqft_lotsquare** -  footage of the lot
* **floorsTotal** -  floors (levels) in house
* **waterfront** - House which has a view to a waterfront
* **view** - Has been viewed
* **condition** - How good the condition is ( Overall )
* **grade** - overall grade given to the housing unit, based on King County grading system
* **sqft_above** - square footage of house apart from basement
* **sqft_basement** - square footage of the basement
* **yr_built** - Built Year
* **yr_renovated** - Year when house was renovated
* **zipcode** - zip
* **lat** - Latitude coordinate
* **long** - Longitude coordinate
* **sqft_living15** - The square footage of interior housing living space for the nearest 15 neighbors
* **sqft_lot15** - The square footage of the land lots of the nearest 15 neighbors

#### 5.0 Dropping unnecessary features and feature creation 

##### 5.1 Dropping the id column and converting date to only years

```python
df['date'] = df['date'].str.slice(0,4).astype(int)
```

```python
df.head()
```

##### 5.2 Converting all the columns data type to float and creating features age

```python
for column in df.columns:
    df[column] = df[column].astype(float)
df['age'] = df['date'] - df['yr_built']
df.head()
```

##### 5.3 Dropping id,date,yr_built,yr_renovated

```python
df = df.drop(['id', 'date', 'yr_built', 'yr_renovated'], axis='columns')
df.head()
```

##### 5.4 Value counts of each column

```python
for column in df:
    print(df[column].value_counts())
    print("*" * 20)
```

#### 6.0 Plotting Correlation plot

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

##### 6.1 Watching in pairwise

```python
corr_features = []
for column,corr_value in df.corr().iterrows():
    k = 0
    for j in range(len(corr_value)):
        if column != corr_value.index[k]:
            if corr_value.values[k] >= 0.5:
                corr_features.append([column, corr_value.index[k], corr_value.values[k]])
        k += 1
corr_features
```

##### 6.2 Watching in respect to Price

```python
df.corr()['price'].sort_values(ascending=False)
```

#### 7.0 Selecting Features

```python
feat =[]
for i in corr_features:
    if i[2] >= 0.8: # 2 is the correlation value as it is shown in the corr_features output above
        feat.append(i[0])
        feat.append(i[1])
df.drop(list(set(feat)), axis=1, inplace=True)
```

#### 9.0 Outlier Detection 

##### 9.1 Histogram and boxplot for price

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
sns.histplot(x=df['price'], ax=axes[0], bins=100, kde=True)
axes[0].set_title('Histogram of Price')
sns.boxplot(data=df, y='price', ax=axes[1])
axes[1].set_title('Boxplot of Price')
plt.show()
```

##### 9.2 Detecting using z-score

```python
from scipy.stats import zscore
z_scores = zscore(df['price'])
outliers_df = pd.DataFrame({'Price': df['price'], 'Z-Score': z_scores})
threshold = 3
outliers = outliers_df[abs(outliers_df['Z-Score']) > threshold]
outliers
```

##### 9.3 Dropping all those outliers

```python
outlier_mask = abs(z_scores) > threshold
df = df[~outlier_mask]
```

```python
df.head()
```

##### 9.4 Watching after removing outliers

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
sns.histplot(x=df['price'], ax=axes[0], bins=100, kde=True)
axes[0].set_title('Histogram of Price')
sns.boxplot(data=df, y='price', ax=axes[1])
axes[1].set_title('Boxplot of Price')
plt.show()
```
